```python
with open('./fnames.txt') as file:
    *fnames, = map(str.strip, file)
```

```python
1/0
chunk_size = 1000
first_chunk = 1
for chunk_ix in tqdm.trange(first_chunk, -(-len(fnames)//chunk_size)):
    start_ix, stop_ix = chunk_ix * chunk_size, (1+chunk_ix) * chunk_size
    
    spaced_fnames_to_archive = ' '.join(map((lambda fname: './some-sound/LibriSpeech-melspecs/'+fname),
                                            fnames[start_ix : stop_ix]))
    archive_name = 'archived/' + just_sha + '-spec-' + str(chunk_ix) + '.tar'
    print(archive_name)
    shell_code = 'tar -cvf {} {}'.format(archive_name, spaced_fnames_to_archive)
    errno = 12332112
    while errno != 0:
        errno = os.system(shell_code)
```
